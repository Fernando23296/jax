```python
#autodiff, primera derivada
def f(x):
    return 3.0*x

df = jax.grad(f)

print(f(2.0))
print(df(2.0))
```

```python
def ff(w,x):
    return w*x

f_jit = jax.jit(ff)
print(ff(3.0, 2.0))
print(f_jit(3.0,2.0))
```
